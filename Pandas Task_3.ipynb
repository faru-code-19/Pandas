# 7. Group the DataFrame by Department and calculate the mean salary for each Department.


```python
grouped=df_employees.groupby('Department')
mean_salary=grouped['Salary'].mean()
print(mean_salary)
```

# 8. Sort the DataFrame by Salary in descending order.

    Sort the DataFrame based on Salary from highest to lowest.


```python
sorted_df=df_employees.sort_values(by='Salary',ascending=False)
print(sorted_df)
```
