# 1. Explain what is Series and DataFrame in Pandas? How do they differ?

Series
- 1D(single array of data)
- 2-Homogeneous data (elements are of same data type)
- 3-series has only 2 objects
         -index and elements/values
- 4-it supports indexing, slicing and are mutable.
- 5-supports indexing for both int and label based indexing.
- 6-supports element wise operation and broadcasting.
- 7-supports statistical operations

DataFrame
- it is a 2D structure
- it consists of rows and columns along with row index and col index
- Heterogeneous data
- it also supports indexing, slicing, and are mutable.
- it supports statistical operations.

# 3. Create a Simple DataFrame using the following dictionary and print it.

Create a DataFrame using the following data:

data = {
    'Name': ['John', 'Marry', 'Steve'],
    'Score': [85, 90, 78],
    'Subject': ['Math', 'English', 'Science']
}


```python
import pandas as pd

data={ 
       'Name': ['John', 'Marry', 'Steve'], 
       'Score': [85, 90, 78], 
       'Subject': ['Math', 'English', 'Science'] 
     }
df_data=pd.DataFrame(data)
df_data
```

# 4. Print only the 'Name' and 'Score' columns from the DataFrame.


```python
df_data[['Name','Score']]
```

# 5. Add a new column called Grade to the DataFrame and assign the value 'A' to all rows.
