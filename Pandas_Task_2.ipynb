- 1- pd.read_csv()

# 10. In a markdown cell, explain the difference between:

to_csv() vs to_excel()

read_csv() vs read_excel()


# to_csv()-
- Saves a DataFrame as a csv(.csv) file
  
Example- 
- df.to_csv("data.csv", index=False)

# read_excel()-
- Reads data from an Excel file(.xlsx or .xls) into a DataFrame
- Syntax-
- pd.read_excel("data.xlsx")

# to_excel()-

- Saves a DataFrame as a Excel(.xlsx) file

- Example-

- df.to_excel("data.xlsx", index=False)


# read_csv()-
- Reads data from an csv file(.csv) into a DataFrame
- Syntax-
- pd.read_csv("data.csv")
