# 1. Replace the department name "HR" with "Human Resources"

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Amit', 'Bhavna', 'Chirag', 'Deepa', 'Eshan', 'Farah'],
    'Department': ['HR', 'Finance', 'IT', 'HR', 'Finance', 'IT'],
    'Experience': [3, 7, 5, 2, 10, 4],
    'Salary': [45000, 70000, 60000, 40000, 85000, 55000]
}

df=pd.DataFrame(data)
df
```

```python
df['Department']=df['Department'].str.replace('HR','Human Resources')
df
```

# 2. Show the name(s) of the highest paid employee(s)

```python
df[df['Salary']==df['Salary'].max()][['Name','Salary']]
```

# 3. Count how many employees are in each department.

```python
df.groupby(by="Department").size().reset_index(name='count of employees')
```

# 4. Create a new column Seniority using experience:
 #### 'Junior' if Experience < 5
 #### 'Mid-level' if Experience is between 5 and 8
 #### 'Senior' if Experience > 8

```python
df["Senioruty"]=np.where(df['Experience']<5,'junior',np.where((df['Experience']>=5) & (df['Experience']<=8),'Mid-level','Senior'))
```

# 5. Find the average salary of only those in the 'IT' department with more than 3 years of experience.


```python
df[ (df['Experience']>3) & (df['Department']=="IT") ]['Salary'].mean()
```
